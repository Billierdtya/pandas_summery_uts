```python
dataset3= pd.read_csv("DDoS UDP Flood.csv") #Membaca file CSV "DDoS UDP Flood.csv" dan menyimpannya dalam variabel `dataset3` sebagai DataFrame.
```

```python
hasilgabung =pd.concat([dataset,dataset2,dataset3],ignore_index=True) #Menggabungkan `dataset`, `dataset2`, dan `dataset3` menjadi satu DataFrame baru bernama `hasilgabung`, dengan mengabaikan indeks asli dan memberikan indeks baru secara berurutan.
```

```python
hasilgabung.columns.values #Menampilkan nama-nama kolom dari DataFrame `hasilgabung` dalam bentuk array.
```

```python
x = hasilgabung.iloc[:,7: 76] #Memilih kolom dari indeks ke-7 hingga ke-75 (kolom 8 sampai 76) dari DataFrame `hasilgabung` dan menyimpannya dalam variabel `x`.
```

```python
x #Menampilkan isi dari variabel `x`, yang berisi subset kolom 8 hingga 76 dari DataFrame `hasilgabung`.
```

```python
y = hasilgabung.iloc[:,83: 84] #Memilih kolom ke-83 hingga ke-83 (kolom 84) dari DataFrame `hasilgabung` dan menyimpannya dalam variabel `y`.
```

```python
y #Menampilkan isi dari variabel `y`, yang berisi kolom ke-84 dari DataFrame `hasilgabung`.
```

```python
from sklearn.model_selection import train_test_split #Mengimpor fungsi `train_test_split` dari `sklearn.model_selection` untuk membagi data menjadi data latih dan data uji.
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42) #Membagi data `x` dan `y` menjadi data latih (80%) dan data uji (20%) secara acak tapi konsisten karena `random_state = 42`.
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier #Mengimpor modul `tree` dari `sklearn` dan kelas `DecisionTreeClassifier` untuk membuat model klasifikasi pohon keputusan.
```

```python
alya = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
alya.fit(x_train,y_train) #Membuat model Decision Tree dengan metode pemisahan 'entropy' dan splitter acak, lalu melatihnya menggunakan data latih `x_train` dan `y_train`.
```

```python
y_pred = alya.predict(x_test) #Memprediksi hasil kelas untuk data uji `x_test` menggunakan model decision tree `alya`, lalu menyimpannya dalam variabel `y_pred`.
```

```python
y_pred #Menampilkan hasil prediksi dari model `alya` terhadap data uji `x_test`.
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred) #Menghitung akurasi prediksi model dengan membandingkan `y_pred` (prediksi) dan `y_test` (label asli), lalu menyimpannya dalam variabel `accuracy`.
```

```python
accuracy #Menampilkan nilai akurasi dari model, yaitu proporsi prediksi yang benar dibanding total data uji.
```

```python
import matplotlib.pyplot as plt
import numpy as np #Mengimpor library `matplotlib.pyplot` untuk visualisasi data (dengan alias `plt`) dan `numpy` untuk operasi numerik (dengan alias `np`).
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(alya, feature_names = x.columns.values, class_names = np.array([ 'Benign Traffic','DDos ICMP Flood','DDoS UDP Flood']), filled = True)
plt.show() #Membuat visualisasi pohon keputusan `alya` dengan ukuran gambar 10x7, menampilkan nama fitur dan kelas, serta mewarnai node berdasarkan kelas, lalu menampilkannya dengan `plt.show()`.
```

```python
import seaborn as lol
from sklearn import metrics
label = np.array([ 'Benign Traffic','DDos ICMP Flood','DDoS UDP Flood']) #Mengimpor `seaborn` dengan alias `lol` untuk visualisasi data dan mengimpor modul `metrics` dari `sklearn` untuk evaluasi model. Baris berikutnya membuat array `label` yang berisi nama kelas untuk digunakan dalam evaluasi atau visualisasi.
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 10))
lol.heatmap(conf_matrix, annot=True, xticklabels=label, yticklabels=label)
plt.xlabel('Prediksi')
plt.ylabel('Fakta')
plt.show() #Membuat dan menampilkan **confusion matrix** menggunakan `seaborn.heatmap` untuk memvisualisasikan perbandingan antara prediksi (`y_pred`) dan fakta asli (`y_test`), dengan label kelas yang sudah didefinisikan dalam variabel `label`.
```
